## ***End-to-End Churn Prediction Pipeline***

---

#### ***INTRODUCTION***

> Customer churn is a major challenge for companies in industries such as telecommunications, where losing customers directly impacts revenue and growth. Churn refers to the situation when customers stop using a company’s services. Therefore, predicting which customers are likely to churn is very important so that companies can take preventive actions.
---
> In this project, an end-to-end machine learning pipeline has been developed using the Telco Customer Churn Dataset to predict customer churn effectively. The main objective of this project is to analyze customer data, identify patterns that lead to churn, and build a predictive model that can estimate the probability of a customer leaving the service.
---
> To achieve this, we first performed data preprocessing and cleaning to handle missing values and convert categorical variables into numerical format. Then, we conducted Exploratory Data Analysis (EDA) to understand the distribution of churn and the relationship between different features such as tenure, monthly charges, and total charges.
---
> After that, feature engineering techniques were applied, such as creating tenure groups and charge ratios, to improve the predictive power of the model. Since the dataset is imbalanced, different sampling techniques like SMOTE, ADASYN, and undersampling were used to handle class imbalance.
---
> Multiple machine learning models, including Logistic Regression, Random Forest, XGBoost, LightGBM, and Support Vector Machine (SVM), were trained and evaluated using Stratified K-Fold Cross-Validation. Hyperparameter tuning was performed using Optuna to improve model performance. The models were evaluated using metrics such as ROC-AUC, Precision-Recall AUC, and F1-score.
---
> Additionally, a cost-sensitive analysis was performed to minimize business loss, and a profit curve was generated to find the optimal decision threshold. SHAP (SHapley Additive exPlanations) was used to interpret the model and identify the key factors influencing churn.
---
> Finally, a prediction function was developed that takes customer data as input and outputs churn probability, key risk factors, and recommended actions to retain the customer.

---

```python
# Import required libraries for data handling, visualization, ML models, and optimization
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import roc_auc_score, f1_score, precision_recall_curve, auc

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC

from xgboost import XGBClassifier
from lightgbm import LGBMClassifier

from imblearn.over_sampling import SMOTE, ADASYN
from imblearn.under_sampling import RandomUnderSampler

import shap
import optuna
```

---

```python
# load data and cleaning
Tele = pd.read_csv("Telecom-Customer-Churn.csv")
Tele.head(2)
```

---

```python
# Drop ID
Tele.drop('customerID', axis=1, inplace=True)
```

---

```python
# Fix TotalCharges
Tele['TotalCharges'] = pd.to_numeric(Tele['TotalCharges'], errors='coerce')
```

---

```python
# Drop missing
Tele.dropna(inplace=True)
```

---

```python
# Target encoding
Tele['Churn'] = Tele['Churn'].map({'Yes':1, 'No':0})
```

---

```python
# Encode ALL categorical columns
cat_cols = Tele.select_dtypes(include=['object']).columns
for col in cat_cols:
    Tele[col] = LabelEncoder().fit_transform(Tele[col].astype(str))
```

---

```python
# EXPLORATORY DATA ANALYSIS (EDA)
print(Tele['Churn'].value_counts(normalize=True))
```

---

```python
for col in ['tenure','MonthlyCharges','TotalCharges']:
    sns.histplot(data=Tele, x=col, hue='Churn', kde=True)
    plt.show()
```

---

```python
# heatmap
plt.figure(figsize=(10,8))
sns.heatmap(Tele.corr(), cmap='coolwarm')
plt.show()
```

---

```python
# FEATURE ENGINEERING
Tele['tenure_group'] = pd.cut(
    Tele['tenure'],
    bins=[0,12,24,48,60,100],
    labels=[0,1,2,3,4]
).astype(int)  
```

---

```python
Tele['charge_ratio'] = Tele['MonthlyCharges']/(Tele['TotalCharges']+1)
```

---

```python
# REMOVE NaN / INF (IMPORTANT)
Tele.replace([np.inf, -np.inf], 0, inplace=True)
Tele.fillna(0, inplace=True)
```

---

```python
# TRAIN-TEST SPLIT
X = Tele.drop('Churn', axis=1)
y = Tele['Churn']
```

---

```python
# Ensure numeric
X = X.astype(float)

X_train, X_test, y_train, y_test = train_test_split(
    X, y, stratify=y, test_size=0.2, random_state=42)
```

---

```python
# Extra safety
X_train = X_train.fillna(0)
X_test = X_test.fillna(0)
```

---

```python
# IMBALANCE HANDLING METHODS
# Define different sampling techniques
methods = {
    "SMOTE": SMOTE(),
    "ADASYN": ADASYN(),
    "Under": RandomUnderSampler()
}
```

---

```python
# MODEL DEFINITIONS
# Define machine learning models
models = {
    "Logistic": LogisticRegression(max_iter=1000),
    "RandomForest": RandomForestClassifier(),
    "XGBoost": XGBClassifier(eval_metric='logloss'),
    "LightGBM": LGBMClassifier(),
    "SVM": SVC(probability=True)
}
```

---


```python
# STRATIFIED K-FOLD
skf = StratifiedKFold(n_splits=5)
results = {}

for m_name, sampler in methods.items():
    X_res, y_res = sampler.fit_resample(X_train, y_train)

    # 🔥 FIX
    X_res = pd.DataFrame(X_res, columns=X_train.columns)
    X_res = X_res.astype(float).fillna(0)

    for name, model in models.items():
        scores = []

        for train_idx, val_idx in skf.split(X_res, y_res):
            X_tr, X_val = X_res.iloc[train_idx], X_res.iloc[val_idx]
            y_tr, y_val = y_res.iloc[train_idx], y_res.iloc[val_idx]

            X_tr = X_tr.fillna(0)
            X_val = X_val.fillna(0)

            model.fit(X_tr, y_tr)
            preds = model.predict_proba(X_val)[:,1]
            scores.append(roc_auc_score(y_val, preds))

        results[(m_name, name)] = np.mean(scores)

print("CV Results:", results)
```

---

```python
# OPTUNA
def objective(trial):
    params = {
        'n_estimators': trial.suggest_int('n_estimators',100,300),
        'max_depth': trial.suggest_int('max_depth',3,8),
        'learning_rate': trial.suggest_float('learning_rate',0.01,0.2)
    }

    model = XGBClassifier(**params, eval_metric='logloss')
    model.fit(X_train, y_train)

    preds = model.predict_proba(X_test)[:,1]
    return roc_auc_score(y_test, preds)

study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=10)

print("Best Params:", study.best_params)
```

---

```python
# FINAL MODEL
model = XGBClassifier(**study.best_params, eval_metric='logloss')
model.fit(X_train, y_train)

preds = model.predict_proba(X_test)[:,1]
```

---

```python
# METRICS
roc = roc_auc_score(y_test, preds)

precision, recall, _ = precision_recall_curve(y_test, preds)
pr_auc = auc(recall, precision)

f1 = f1_score(y_test, (preds>0.5).astype(int))

print("ROC:", roc)
print("PR-AUC:", pr_auc)
print("F1:", f1)
```

---

```python
# COST + PROFIT CURVE
def cost(y_true, prob, t):
    pred = (prob>t).astype(int)
    fn = ((y_true==1)&(pred==0)).sum()
    fp = ((y_true==0)&(pred==1)).sum()
    return fn*500 + fp*50

thresholds = np.linspace(0,1,50)
costs = [cost(y_test, preds, t) for t in thresholds]

plt.plot(thresholds, costs)
plt.title("Profit Curve")
plt.show()
```

---

```python
# SHAP
explainer = shap.Explainer(model)
shap_values = explainer(X_test)

shap.summary_plot(shap_values, X_test)
```

---

```python
# FINAL PREDICTION FUNCTION
def predict_customer(customer):
    customer = customer.astype(float).fillna(0)

    prob = model.predict_proba(customer)[:,1][0]
    shap_val = explainer(customer)

    impact = np.abs(shap_val.values).mean(axis=0)
    top = customer.columns[np.argsort(impact)[-3:]]

    actions=[]
    for f in top:
        if "Charges" in f:
            actions.append("Offer discount")
        elif "tenure" in f:
            actions.append("Loyalty rewards")
        else:
            actions.append("Engagement")

    return {
        "probability":prob,
        "top_factors":list(top),
        "actions":actions
    }

print(predict_customer(X_test.iloc[[0]]))
```
